```python
import pandas as pd
import re
from bs4 import BeautifulSoup
import requests
```

```python
URL_2003 ='https://en.wikipedia.org/wiki/2003_Rugby_World_Cup'
response_2003 = requests.get(URL_2003)
soup_2003 = BeautifulSoup(response_2003.text,'html.parser')
div_2003 = soup_2003.find('div', {'id':'mw-content-text'})
summary_2003 = div_2003.find_all('div',{'class':'vevent summary'})
```

```python
events_2003 = []
for i in summary_2003:
    text = i.text
    events_2003.append(text)
```

```python
events_clean_2003 = [e.replace('\n',' ').replace('\xa0', ' ').strip() for e in events_2003]
data_2003 = " ".join(events_clean_2003)
```

```python
data_2003 = data_2003.replace('(a.e.t.)','')
data_2003 = data_2003.replace('-','–')
```

```python
#datemethodregex
date_pattern = re.compile(r'\d{1,2}\s\w*\s2003')
dates = date_pattern.findall(data_2003)
date_list = []
for i in dates:
    date_list.append(i)
```

```python
#team1
team1_pattern = re.compile(r'\s{4,6}(\D*|\D*\s\D*)\s{2,3}\d{1,3}–')
team1 = team1_pattern.findall(data_2003)
team1_list = []
for i in team1:
    team1_list.append(i)
```

```python
#team2
team2_pattern = re.compile(r'[–]\d*\s{3,4}(\w*|\w*\s\w*)\s{3,7}\D{3}:')
team2 = team2_pattern.findall(data_2003)
team2_list = []
for i in team2:
    team2_list.append(i)
```

```python
#score
score_pattern = re.compile(r'\d*[–]\d*')
scores = score_pattern.findall(data_2003)
scores_list = []
for i in scores:
    scores_list.append(i)
```

```python
#points team 1
pt1_pattern = re.compile(r'\s{2,4}(\d*)[–]\d*\s{2,4}')
pt1 = pt1_pattern.findall(data_2003)
pt1_list = []
for i in pt1:
    pt1_list.append(i)
pt1_list = [ int(x) for x in pt1_list ]

#points team 2 
pt2_pattern = re.compile(r'\s{2,4}\d*[–](\d*)\s{2,4}')
pt2 = pt2_pattern.findall(data_2003)
pt2_list = []
for i in pt2:
    pt2_list.append(i)
pt2_list = [ int(x) for x in pt2_list ]

#total points
total_points_list =[]
for i,j in zip(pt1_list, pt2_list):
    k = i+j
    total_points_list.append(k)
```

```python
#attendance
attendance_pattern = re.compile(r'Attendance:\s(\d*[,]\d*)')
attendance = attendance_pattern.findall(data_2003)
attendance_list = []
for i in attendance:
    attendance_list.append(i)
```

```python
#referee
referee_pattern = re.compile(r'Referee:\s(\w*\s\w*)')
referee = referee_pattern.findall(data_2003)
referee_list = []
for i in referee:
    referee_list.append(i)
```

```python
#referee nationality
ref_nationality_pattern = re.compile(r'[(](\D*\s\D*|\D*)[)]')
ref_nationality = ref_nationality_pattern.findall(data_2003)
referee_nationality_list = []
for i in ref_nationality:
    referee_nationality_list.append(i)
```

```python
dico_2003 = {'Match Date':date_list,'Home Team': team1_list, 'Away Team' : team2_list, 'Score': scores_list, 'Points Home Team': pt1_list, 'Points Away Team' : pt2_list, 'Total Points': total_points_list, 'Attendance': attendance_list, 'Referee': referee_list, 'Referee Nationality': referee_nationality_list}
df_2003 = pd.DataFrame.from_dict(dico_2003, orient='index')
df_2003 = df_2003.transpose()
df_2003
```
